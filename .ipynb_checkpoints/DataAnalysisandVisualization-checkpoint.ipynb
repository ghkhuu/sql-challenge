```python
from sqlalchemy import create_engine
import pymysql
import pandas as pd

db_connection_str = 'mysql+pymysql://root:iamroot@localhost/sql_challenge'
db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT e.emp_no, e.first_name, e.last_name, e.gender, s.salary, t.title \
                    FROM Employees e LEFT JOIN Salaries s ON e.emp_no=s.emp_no LEFT JOIN Titles t \
                    ON e.emp_no=t.emp_no',
                 con=db_connection)
df.head(10)
```
