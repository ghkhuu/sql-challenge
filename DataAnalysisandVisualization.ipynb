```python
from sqlalchemy import create_engine
import pymysql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (18,6)
```

```python
db_connection_str = 'mysql+pymysql://root:iamroot@localhost/sql_challenge'
db_connection = create_engine(db_connection_str)

# Custom SQL to extract desired result set from MySQL database
df = pd.read_sql('SELECT e.emp_no, e.first_name, e.last_name, e.gender, s.salary, t.title, d.dept_name \
                    FROM Employees e \
                    LEFT JOIN Salaries s ON e.emp_no=s.emp_no \
                    LEFT JOIN Titles t ON e.emp_no=t.emp_no \
                    LEFT JOIN (SELECT * FROM Dept_emp UNION ALL SELECT * FROM Dept_manager) t1 ON t1.emp_no=e.emp_no \
                    LEFT JOIN Departments d ON t1.dept_no=d.dept_no',
                 con=db_connection)
df.head(10)
```

```python
# Making a histogram to visualize the most common salary ranges for employees. Based on the data, a majority of employees
# earns between $40-50k

n_bin = 10
x = df["salary"]
plt.hist(x, bins=n_bin, color='teal')
```

```python
# Aggregate data to calculate mean across different attributes (job title, gender, department)
avgSalaryByTitle = df[["title","salary"]].groupby(["title"]).mean()
avgSalaryByGender = df[["gender","salary"]].groupby(["gender"]).mean()
avgSalaryByDepartment = df[["dept_name","salary"]].groupby(["dept_name"]).mean()

# Creating figure object with multiple subplots
figure, axes = plt.subplots(nrows=1, ncols=3, sharey=True)
figure.suptitle('Average Salary across Different Attributes', size=16)


# Adding axes to each subplots
axes[0].bar(x=avgSalaryByTitle.index, height=avgSalaryByTitle['salary'], width=0.8)
axes[0].set_title('by Job Title')
axes[1].bar(x=avgSalaryByGender.index, height=avgSalaryByGender['salary'], width=0.8)
axes[1].set_title('by Gender')
axes[2].bar(x=avgSalaryByDepartment.index, height=avgSalaryByDepartment['salary'], width=0.8)
axes[2].set_title('by Department')

# Rotating tick labels
for tick in axes[0].get_xticklabels():
    tick.set_rotation(270)
for tick in axes[2].get_xticklabels():
    tick.set_rotation(270)
```

```python
# Saving dataframe to .csv for further visualizations
df.to_csv('sql_challenge_extract.csv', index_label='index')
```
